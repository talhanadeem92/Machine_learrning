```python
import pandas as pd
```

```python
data=pd.read_csv("/content/car data.csv")
```

```python
data.describe()
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average selling price for each fuel type
avg_selling_price = data.groupby('Fuel_Type')['Selling_Price'].mean().reset_index()

# Correct Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Fuel_Type', y='Selling_Price', data=avg_selling_price, palette=['blue', 'gray', 'green'])
plt.title('Average Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Selling Price')
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate manipulated data for deceptive visualization
np.random.seed(0)
kms_driven = data['Kms_Driven']
selling_price =data['Selling_Price'] + np.random.normal(0, 2, len(data))

# Deceptive Visualization
plt.figure(figsize=(10, 6))
plt.scatter(kms_driven, selling_price, s=50, alpha=0.5, color='orange')
plt.title('Relationship between Kms Driven and Selling Price')
plt.xlabel('Kms Driven')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
```
