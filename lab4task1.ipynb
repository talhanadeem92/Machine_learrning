**2.pandas_profiling**

**The purpose of pandas_profiling in simple terms is to automatically generate a detailed report about a dataset, providing valuable insights and statistics, which helps data analysts and scientists quickly understand and assess the data's characteristics and quality.**

```python
pip install pandas-profiling
```

```python
import pandas as pd
from pandas_profiling import ProfileReport

# Create a sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [5.1, 6.2, 7.3, 8.4, 9.5],
        'C': ['apple', 'banana', 'cherry', 'date', 'elderberry']}
df = pd.DataFrame(data)

# Create a profiling report
profile = ProfileReport(df)

# Generate and save the HTML report
profile.to_file("report.html")
```
