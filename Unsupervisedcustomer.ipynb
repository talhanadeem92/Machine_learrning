```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming 'data' contains your preprocessed dataset

# Select the features you want to use for clustering
features_for_clustering = ['Number_of_houses', 'Avg_size_household', 'Avg_age']

# Choose the number of clusters (you can adjust this based on your data)
n_clusters = 6

# Create a K-Means model
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Fit the model to the selected features
data['Cluster'] = kmeans.fit_predict(data[features_for_clustering])

# Visualize the clusters (2D plot for two features)
plt.figure(figsize=(8, 6))
plt.scatter(data[features_for_clustering[0]], data[features_for_clustering[1]], c=data['Cluster'], cmap='viridis')
plt.xlabel(features_for_clustering[0])
plt.ylabel(features_for_clustering[1])
plt.title('K-Means Clustering')
plt.show()

# You can repeat this process for different combinations of features and clusters as needed.
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(data[features_for_clustering], data['Cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```

```python
from sklearn.impute import SimpleImputer

# Assuming 'data' contains your preprocessed dataset
# Specify the strategy for imputation (e.g., 'mean', 'median', 'constant')
imputer = SimpleImputer(strategy='mean')

# Select the features to impute (features_for_clustering in this case)
data[features_for_clustering] = imputer.fit_transform(data[features_for_clustering])

# Now, your dataset no longer contains missing values, and you can proceed with K-Means clustering.
```
