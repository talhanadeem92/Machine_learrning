```r

mpg <- mtcars$mpg

# Load the MASS package for the fitdistr() function
library(MASS)

# Fit a normal distribution to the data
fit <- fitdistr(mpg, "normal")


# Generate the density values based on the fitted normal distribution
x <- seq(min(mpg), max(mpg), length.out = 100)
density_values <- dnorm(x, mean = fit$estimate[1], sd = fit$estimate[2])

# Plot the histogram of the data
hist(mpg, freq = FALSE, main = "Histogram of MPG in mtcars Dataset", xlab = "MPG", ylab = "Density")

# Add the fitted density curve to the histogram plot
lines(x, density_values, col = "red", lwd = 2)

# Add a legend
legend("topright", legend = c("Fitted Normal Distribution"), col = "red", lty = 1, lwd = 2)
```

```r
# Load the mtcars dataset if you haven't already done so
data(mtcars)

# Extract the variable of interest (e.g., wt)
wt <- mtcars$wt

# Fit an exponential distribution to the data
fit_exp <- fitdistr(wt, "exponential")

# Extract the parameter of the fitted distribution (rate parameter)
rate <- fit_exp$estimate

# Create a histogram of the data
hist(wt, breaks = 10, freq = FALSE, main = "Histogram of Weight (wt) in mtcars Dataset", xlab = "Weight (wt)", ylab = "Density")

# Generate density values based on the fitted exponential distribution
x <- seq(min(wt), max(wt), length.out = 100)
density_values_exp <- dexp(x, rate = rate)

# Add the fitted density curve to the histogram plot
lines(x, density_values_exp, col = "blue", lwd = 2)
```

```r
# Extract the variable of interest (e.g., disp)
disp <- mtcars$disp

# Perform Kernel Density Estimation (KDE)
dens <- density(disp)

# Plot the estimated density
plot(dens, main = "Kernel Density Estimation of Engine Displacement (disp) in mtcars Dataset",
     xlab = "Engine Displacement (disp)", ylab = "Density", col = "blue", lwd = 2)

# Add a rug plot to show individual data points along the x-axis
rug(disp, col = "red")
```
